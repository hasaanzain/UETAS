```python
import pandas as pd 
df = pd.read_csv('uetas_qasim.csv')
```

```python
names = df['Full Name']
names_clean = names.str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
names_split = names_clean.str.split(' ')
first_names = names_split.str[0]
last_names = names_split.str[-1]
```

```python
js_objects = [
    f"    {{ firstName: '{fn}', lastName: '{ln}' }}"
    for fn, ln in zip(first_names, last_names)
]

# Build final JS output
js_output = (
    "const registeredAttendees = [\n"
    + ",\n".join(js_objects)
    + "\n];"
)

print(js_output)
```
